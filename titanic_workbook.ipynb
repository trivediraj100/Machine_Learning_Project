```python
from sklearn import datasets
import numpy as np
import pandas as pd
data1 = pd.read_csv("titanic_train.csv")
# data_1_new = data1.drop(columns = {"Name", "Ticket"} , axis = 1)
```

```python
#changing the name column

name = []
for i in data1['Name']:
    name.append(i.split(",")[1].split(".")[0].strip())
name = np.array(name)
name[name == "Master"] = 0
name[name == "Miss"] = 1
name[name == "Mr"] = 2
name[name == "Mrs"] = 3
name[(name!='0')&(name!='1')&(name!='2')&(name!='3')]=4
name = np.array(name , dtype = int)
```

```python
data1 = data1.drop(columns = {"Name", "Ticket"} , axis = 1)
```

```python
data1 = data1.drop(columns = {"Cabin"} , axis = 1)

data1['name'] = name
data1
```

```python
sex = []
for i in data1['Sex']:
    if i == "female":
        sex.append(0)
    else:
        sex.append(1)
sex = np.array(sex)
sex
data1["sex"] = sex
```

```python
data1.drop(columns = "Sex", inplace = True)
```

```python
data1.loc[data1["Embarked"].isnull() , "Embarked"] = "S"
data1.loc[data1["Embarked"] == "S" , "Embarked"] = "1"
data1.loc[data1["Embarked"] == "C" , "Embarked"] = "2"
data1.loc[data1["Embarked"] == "Q" , "Embarked"] = "3"
data1.loc[data1["Embarked"] == "" , "Embarked"] = "1"

data1.loc[data1["Age"].isnull(), "Age"] = data1.loc[data1["Age"].notnull(), 'Age' ].median()
emabrked = np.array(data1['Embarked'] , dtype = float)
survived = []
for i in data1['Survived']:
    survived.append(i)
survived = np.array(survived)
data1["survived"] = survived
data1.drop(columns = {"Survived"} , inplace = True)
```
